```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([1,2,3,4], dtype=torch.float32)
Y = torch.tensor([2,4,6,8], dtype=torch.float32)

w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
```

####Training Pipeline: Model, Loss, and Optimizer


1.   Design model (input, output, forward, layers)
2.   Construct loss and optimizer 
3.   Training loop
  *   Forward pass: Compute preductin 
  *   Backward pass: Gradient
  *   Update Weights

```python
# Model Prediction

def forward(x):
  return w*x
```

```python
print("Prediction before training: f(5) = {:.3f}".format(forward(5)))
```

```python
learning_rate = 0.01
n_iters = 100 # epochs 

# Defining loss before iterating 
loss = nn.MSELoss()
optimizer = torch.optim.SGD([w], lr =learning_rate )

for epoch in range(n_iters):
  # prediction = forward pass
  y_pred = forward(X)

  # loss
  l = loss(Y, y_pred)

  # gradients = backward pass
  l.backward() # dl/dw

  # update weights 
  optimizer.step()

  # zero our gradients 
  optimizer.zero_grad()


  n = epoch +1
  if epoch % 5 == 0:
    print("Epoch:{} Weight:{:.3f} Loss:{:.4f}".format(n, w, l)) 
```

```python
print("Prediction after training: f(5) = {:.3f}".format(forward(5)))
```
