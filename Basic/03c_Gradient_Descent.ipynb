```python
import torch
import torch.nn as nn
```

#### Full Pytorch Simple Linear Regression

```python
# Adding shape to the data
X = torch.tensor([[1],[2],[3],[4]], dtype=torch.float32)
Y = torch.tensor([[2],[4],[6],[8]], dtype=torch.float32)

n_samples, n_features = X.shape
print(n_samples)
print(n_features)
#X = torch.tensor([1,2,3,4], dtype=torch.float32)
#Y = torch.tensor([2,4,6,8], dtype=torch.float32)
#w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True) no longer needed
```

```python
# Model Prediction

"""def forward(x):
  return w*x"""
# Define model
input_size = n_features
output_size = n_features
# nn.Linear only provided one layer
model = nn.Linear(input_size, output_size)
```

```python
# Must have the same shape of model
X_test = torch.tensor([5], dtype=torch.float32)

#print("Prediction before training: f(5) = {:.3f}".format(forward(5)))
print("Prediction before training: f(5) = {:.3f}".format(model(X_test).item()))
```

```python
learning_rate = 0.01
n_iters = 100 # epochs 

# Defining loss before iterating 
loss = nn.MSELoss()
# edit weights 
optimizer = torch.optim.SGD(model.parameters(), lr =learning_rate)

for epoch in range(n_iters):
  # prediction = forward pass
  #y_pred = forward(X)
  y_pred = model(X)

  # loss
  l = loss(Y, y_pred)

  # gradients = backward pass
  l.backward() # dl/dw

  # update weights 
  optimizer.step()

  # zero our gradients 
  optimizer.zero_grad()


  n = epoch +1
  if epoch % 5 == 0:
    # Unpack weights 
    [w, b] = model.parameters()
    print("Epoch:{} Weight:{:.3f} Loss:{:.4f}".format(n, w[0][0].item(), l)) 
```
